```python
import cv2
from ultralytics import YOLO
import torch
import numpy as np

# Load the model
#model = YOLO('/path/to/your/model.pt')  # Ensure the correct path to your model

# Open the input video
video_path = r"/home/eigenmaps-01/Downloads/Poultry/1.mp4"
cap = cv2.VideoCapture(video_path)

# Get the video details
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
output_path = r'/home/eigenmaps-01/Downloads/Poultry/path_to_output_video.avi'

# Initialize the video writer
out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'XVID'), fps, (frame_width, frame_height))

# Initialize VideoSink for real-time streaming
sink = cv2.VideoWriter('appsrc ! videoconvert ! x264enc tune=zerolatency ! rtph264pay ! udpsink host=127.0.0.1 port=5000',
                       cv2.CAP_GSTREAMER, 0, fps, (frame_width, frame_height), True)

# Process the video frame by frame
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Make predictions on the frame
    results = model(frame)

    # Extract the results
    boxes = results[0].boxes.xyxy.cpu().numpy()
    masks = results[0].masks.data.cpu().numpy() if results[0].masks is not None else []

    # Draw the instance segmentation masks and bounding boxes
    for box, mask in zip(boxes, masks):
        x1, y1, x2, y2 = map(int, box)

        # Draw bounding box
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # Draw mask
        mask = (mask[0] > 0.5).astype(np.uint8) * 255

        # Resize mask to match frame size
        mask_resized = cv2.resize(mask, (frame_width, frame_height))
        
        # Convert mask to 3 channels
        colored_mask = cv2.applyColorMap(mask_resized, cv2.COLORMAP_JET)

        # Combine frame and mask
        frame = cv2.addWeighted(frame, 1, colored_mask, 0.5, 0)

    # Write the frame to the output video
    out.write(frame)

    # Stream the frame using VideoSink
    sink.write(frame)

# Release resources
cap.release()
out.release()
sink.release()

print("Video processing and streaming completed.")
```

```python
import cv2
from ultralytics import YOLO
import numpy as np
from sort import Sort  # Adjust the import as needed

# Load the YOLO model
model = YOLO('/path/to/your/model.pt')  # Ensure the correct path to your model

# Open the input video
video_path = r"/home/eigenmaps-01/Downloads/Poultry/1.mp4"
cap = cv2.VideoCapture(video_path)

# Get the video details
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
output_path = r'/home/eigenmaps-01/Downloads/Poultry/path_to_output_video3.avi'

# Initialize the video writer
out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'XVID'), fps, (frame_width, frame_height))

# Initialize VideoSink for real-time streaming
sink = cv2.VideoWriter('appsrc ! videoconvert ! x264enc tune=zerolatency ! rtph264pay ! udpsink host=127.0.0.1 port=5000',
                       cv2.CAP_GSTREAMER, 0, fps, (frame_width, frame_height), True)

# Initialize SORT tracker
tracker = Sort()

# Process the video frame by frame
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Make predictions on the frame
    results = model(frame)

    # Extract the results
    boxes = results[0].boxes.xyxy.cpu().numpy()

    # Prepare bounding boxes for the tracker (format: [x1, y1, x2, y2])
    detections = []
    for box in boxes:
        x1, y1, x2, y2 = map(int, box)
        detections.append([x1, y1, x2, y2])

    # Update the tracker with the new detections
    trackers = tracker.update(np.array(detections))

    # Draw the bounding boxes and assign IDs
    for tracker in trackers:
        x1, y1, x2, y2, obj_id = map(int, tracker)
        
        # Draw bounding box
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # Draw object ID
        label = f"ID: {obj_id}"
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Write the frame to the output video
    out.write(frame)

    # Stream the frame using VideoSink
    sink.write(frame)

# Release resources
cap.release()
out.release()
sink.release()

print("Video processing and streaming completed.")
```
